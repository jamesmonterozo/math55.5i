```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from fractions import Fraction
```

```python
n = 20
nums = list()
sums = list()
rSum = 0

for i in range(1, n + 1):
    nums.append(Fraction(i, np.math.factorial(i + 1)))
    rSum += Fraction(i, np.math.factorial(i + 1))
    sums.append(rSum)
```

```python
df = pd.DataFrame(list(zip(range(1, n+1), nums, sums)), 
                    columns = ['n', 'nth Term', 'sum at n'])

df
```

```python
df_converted = df.astype({"nth Term":"float", "sum at n":"float"})
df_converted.head(5)
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(7, 7))

sns.lineplot(x="n", y="nth Term", ax=ax0, data=df_converted)
ax0.set(xlabel="n", ylabel="nth Term", title="The nth Term")

sns.lineplot(x="n", y="sum at n", ax=ax1, data=df_converted)
ax1.set(xlabel="n", ylabel="sum at n", title="Sum at n")

plt.xticks(range(1, n+1))
plt.show()
```
